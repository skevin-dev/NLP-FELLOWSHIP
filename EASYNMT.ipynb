```python
from easynmt import EasyNMT
import json
import pandas as pd
```

```python
model  = EasyNMT('opus-mt')
```

```python
sentences = ['shyaka likes to eat spice food','arnaud is tall','Macron is president of france']
```

```python
translations = model.translate(sentences,target_lang='rw')
```

```python
translations
```

```python
with open('/content/languages.json','r+') as file:
  content =  file.read()
```

```python
languages_dict = json.loads(content)
```

```python
codes = [languages_dict[index]['code'] for index in range(len(languages_dict))]
r = codes[0:5]
```

```python
df = pd.DataFrame()
```

```python
for code in r:
  try: 
    df[code] = model.translate(sentences,target_lang = code)
  except:
    pass 
```

```python
df['sentences'] = sentences
```
